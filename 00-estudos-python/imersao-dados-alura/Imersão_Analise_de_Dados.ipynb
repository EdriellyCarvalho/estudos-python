## Demonstrao de como preencher valores nulos com uma mdia

```python
import numpy as np
## criao de uma base de teste para exemplo de substituio de nulos por uma mdia

df_salarios = pd.DataFrame({
    "nome": ["Amlia", "Junior", "Donatela", "Ariel", "Val"],
    "salario": [4000, np.nan, 5000, np.nan, 1000000]
})

## calcula a mdia e substitui os nulos arredodando as casas decimais
df_salarios["salario_media"] = df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2))

## calcula a mediana e substitui os nulos pela mediana
df_salarios["salario_mediana"] = df_salarios["salario"].fillna(df_salarios["salario"].median())
df_salarios
```

```python
## substituio de valores nulo pelo valor anterior ou posterior

df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda-feira", "Tera-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

## ffill -> completa com o valor anterior
## bfill -> completa com o valor posterior

df_temperaturas["Preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas["Preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
## Substituio de valores nulos por algum valor fixo

df_cidades = pd.DataFrame({
    "nome": ["Amlia", "Junior", "Donatela", "Ariel", "Val"],
    "cidade": ["So Paulo", np.nan, "Curitiba", np.nan, "Belm"]
})

df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Desconhecida")
df_cidades
```

```python
## DESAFIO -  Grfico de salrio mdio por pas para o cargo de Data Science - Verso 1

df_desafio_limpo = df.dropna(subset=["salario", "residencia", "cargo"])
df_ds = df_desafio_limpo["cargo"].str.contains("Data Scientist", case=False)
```

```python
# Filter for 'Data Analyst' roles based on employee residence
df_da_filtered = df_desafio_limpo[df_desafio_limpo["cargo"].str.contains("Data Analyst", case=False)].copy()

# Calculate the average salary in USD per employee residence
salario_medio_residencia_usd = df_da_filtered.groupby("residencia")["usd"].mean().reset_index()

# Display the head of the resulting DataFrame
display(salario_medio_residencia_usd.head())
```

```python
# Calculate the overall average salary in USD for Data Analysts
overall_avg_usd_da = df_da_filtered["usd"].mean()

# Prepare tooltip data
tooltip_data_da = {}
for index, row in salario_medio_residencia_usd.iterrows():
    residencia = row['residencia']
    usd_salary = row['usd']

    # Calculate the percentage relative to the overall average
    percentage_of_avg = (usd_salary / overall_avg_usd_da) * 100 if overall_avg_usd_da != 0 else 0

    tooltip_data_da[residencia] = f"Pas de Residncia: {residencia}<br>Mdia Salarial (USD): {usd_salary:.2f}<br>Porcentagem da Mdia Global: {percentage_of_avg:.2f}%<br>"

# Convert tooltip_data to a list of dictionaries for merging
tooltip_list_da = [{"residencia": country, "tooltip": data} for country, data in tooltip_data_da.items()]
tooltip_df_da = pd.DataFrame(tooltip_list_da)

# Merge the tooltip data with the average USD salary data
salario_mapa_data_da = pd.merge(salario_medio_residencia_usd, tooltip_df_da, on="residencia")

# Display the head of the merged data
display(salario_mapa_data_da.head())
```

```python
# Create a mapping from ISO 2-letter codes to ISO 3-letter codes for the countries in the data
# This is a partial mapping based on the countries observed in the data.
# A more comprehensive mapping could be used if needed.
country_code_mapping = {
    'AD': 'AND', 'AE': 'ARE', 'AR': 'ARG', 'AT': 'AUT', 'AU': 'AUS', 'BE': 'BEL',
    'BG': 'BGR', 'BR': 'BRA', 'CA': 'CAN', 'CH': 'CHE', 'CL': 'CHL', 'CZ': 'CZE',
    'DE': 'DEU', 'DZ': 'DZA', 'EC': 'ECU', 'EE': 'EST', 'EG': 'EGY', 'ES': 'ESP',
    'FI': 'FIN', 'FR': 'FRA', 'GB': 'GBR', 'GR': 'GRC', 'HK': 'HKG', 'HU': 'HUN',
    'ID': 'IDN', 'IE': 'IRL', 'IL': 'ISR', 'IN': 'IND', 'IT': 'ITA', 'JP': 'JPN',
    'KE': 'KEN', 'KR': 'KOR', 'LT': 'LTU', 'LU': 'LUX', 'LV': 'LVA', 'MX': 'MEX',
    'MY': 'MYS', 'NG': 'NGA', 'NL': 'NLD', 'NO': 'NOR', 'NZ': 'NZL', 'PK': 'PAK',
    'PL': 'POL', 'PT': 'PRT', 'RO': 'ROU', 'RU': 'RUS', 'SE': 'SWE', 'SG': 'SGP',
    'SK': 'SVK', 'TH': 'THA', 'TR': 'TUR', 'TW': 'TWN', 'UA': 'UKR', 'US': 'USA',
    'ZA': 'ZAF'
}

# Add a new column with ISO 3-letter codes to the DataFrame
salario_mapa_data_da['residencia_iso3'] = salario_mapa_data_da['residencia'].map(country_code_mapping)

# Display the head of the DataFrame with the new column
display(salario_mapa_data_da.head())
```

```python
import plotly.express as px

# Create the interactive choropleth map using Plotly Express
fig = px.choropleth(salario_mapa_data_da,
                    locations="residencia_iso3", # Column with ISO-3 country codes
                    locationmode="ISO-3", # Set to use ISO-3 codes
                    color="usd", # Column to color the map
                    hover_name="residencia", # Column to display on hover
                    hover_data={"residencia": False, "usd": False, "tooltip": True}, # Include tooltip data, hide default hover data
                    title="Mdia Salarial para Data Analysts por Pas de Residncia (USD)",
                    color_continuous_scale=px.colors.sequential.Plasma, # Color scale
                    projection="natural earth" # Map projection
                   )

# Update layout for better appearance
fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular' # You can experiment with different projections
    ),
    margin={"r":0,"t":40,"l":0,"b":0} # Adjust margins
)

# Update traces to use the custom tooltip
fig.update_traces(hovertemplate='%{customdata[0]}')


# Show the map
fig.show()
```

```python

import pycountry
```

```python
# Funo para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None
```

```python
# Criar nova coluna com cdigo ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)
```

```python
# Calcular mdia salarial por pas (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()
```

```python
# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='RdBu',
                    title='Salrio mdio de Cientista de Dados por pas',
                    labels={'usd': 'Salrio mdio (USD)', 'residencia_iso3': 'Pas'})

fig.show()
```
